```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

subjects = ['maths','english','computer','urdu','chemistry','physics','arabic']
marks = [100, 99, 98, 97, 96, 95, 99]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x_pos = range(len(subjects))
y_pos = [0] * len(subjects)
z_pos = [0] * len(subjects)

dx = [0.6] * len(subjects)
dy = [0.6] * len(subjects)
dz = marks

# draw bars & enable picking
bars = ax.bar3d(x_pos, y_pos, z_pos, dx, dy, dz, picker=True)

ax.set_xticks(list(x_pos))
ax.set_xticklabels(subjects, rotation=20)
ax.set_zlabel("Marks")
ax.set_title("Students Marks (Interactive 3D Chart)")

highlight_label = None
highlight_bar = None

def on_pick(event):
    global highlight_label, highlight_bar

    bar = event.artist
    index = list(bars).index(bar)

    # reset previous highlight
    if highlight_bar:
        highlight_bar.set_color("C0")
    if highlight_label:
        highlight_label.remove()

    # highlight clicked bar
    bar.set_color("orange")
    highlight_bar = bar

    # show value text above bar
    highlight_label = ax.text(
        x_pos[index],
        0,
        marks[index] + 2,
        f"{subjects[index]} = {marks[index]}",
        fontsize=10,
        weight="bold"
    )

    fig.canvas.draw_idle()

fig.canvas.mpl_connect("pick_event", on_pick)

plt.show()
```

```python
employees = ["Ahmad", "Ali", "Hamza"]
salaries = [50000, 45000, 52000]

# Apply 10% increase
updated_salaries = [s * 1.10 for s in salaries]

# Plot line graph
plt.figure(figsize=(8,5))

# Old salaries (red dashed line)
plt.plot(employees, salaries, marker='o', linestyle='--', color='red', label='Old Salary')
# Add salary labels on old salaries
for i, salary in enumerate(salaries):
    plt.text(employees[i], salary + 500, f"{salary}", ha='center', color='red')

# Updated salaries (green solid line)
plt.plot(employees, updated_salaries, marker='o', linestyle='--', color='green', label='Updated Salary')
# Add salary labels on updated salaries
for i, salary in enumerate(updated_salaries):
    plt.text(employees[i], salary + 500, f"{int(salary)}", ha='center', color='green')

plt.title("Old vs Updated Salaries")
plt.xlabel("Employees")
plt.ylabel("Salary")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("AirQuality.csv", sep=";")

print(df.head())
print(df.shape)
print(df.info())
print(df.isnull().sum())
print(df.describe())

df["datetime"] = pd.to_datetime(
    df["Date"].astype(str) + " " + df["Time"].astype(str),
    dayfirst=True,
    errors="coerce"
)

df = df.fillna(df.mean(numeric_only=True))

df["Year"] = df["datetime"].dt.year
df["Month"] = df["datetime"].dt.month
df["Day"] = df["datetime"].dt.day
df["Hour"] = df["datetime"].dt.hour

df["AQI"] = df["NO2(GT)"]

def pollution_level(aqi):
    if aqi <= 50:
        return "Good"
    elif aqi <= 100:
        return "Moderate"
    elif aqi <= 150:
        return "Unhealthy for Sensitive"
    elif aqi <= 200:
        return "Unhealthy"
    elif aqi <= 300:
        return "Very Unhealthy"
    else:
        return "Hazardous"

df["Pollution_Level"] = df["AQI"].apply(pollution_level)

print(df.head())
print(df["Pollution_Level"].value_counts())

monthly_avg = df.groupby("Month")["AQI"].mean()
print(monthly_avg)

plt.figure(figsize=(6,4))
sns.histplot(df["AQI"], kde=True)
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(x=df["AQI"])
plt.show()

plt.figure(figsize=(7,4))
monthly_avg.plot(marker="o")
plt.show()

plt.figure(figsize=(7,4))
sns.countplot(x="Pollution_Level", data=df)
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.show()
```

```python
class BankAccount:
    def __init__(self, user_name, account_number, balance):
        self.user_name = user_name
        self.account_number = account_number
        self.balance = balance
        self.total_deposit = 0
        self.total_withdraw = 0
    def display_account(self):
        print("User Name     :", self.user_name)
        print("Account No   :", self.account_number)
        print("Balance      : Rs.", self.balance)
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            self.total_deposit += amount
            print("Deposited: Rs.", amount)
        else:
            print("Invalid deposit amount")
    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient balance")
        elif amount <= 0:
            print("Invalid withdraw amount")
        else:
            self.balance -= amount
            self.total_withdraw += amount
            print("Withdrawn: Rs.", amount)
    def show_summary(self):
        print("User Name       :", self.user_name)
        print("Account Number :", self.account_number)
        print("Total Deposited: Rs.", self.total_deposit)
        print("Total Withdrawn: Rs.", self.total_withdraw)
        print("Final Balance  : Rs.", self.balance)
name = input("Enter your name: ")
acc_no = input("Enter your account number: ")
bal = float(input("Enter opening balance: "))

account = BankAccount(name, acc_no, bal)

while True:
    print("Choose an option:")
    print("1. Display Account")
    print("2. Deposit Money")
    print("3. Withdraw Money")
    print("4. Show Summary & Exit")

    choice = input("Enter choice (1-4): ")

    if choice == "1":
        account.display_account()
    elif choice == "2":
        amount = float(input("Enter deposit amount: "))
        account.deposit(amount)
    elif choice == "3":
        amount = float(input("Enter withdraw amount: "))
        account.withdraw(amount)
    elif choice == "4":
        account.show_summary()
        print("Thank you for using our bank system")
        break
    else:
        print("Invalid choice, Try again.")
```

```python
class FastFoodOrderMenu:
    def __init__(self):
        self.menu = {
            1: {"name": "Chicken Pizza", "sizes": {"Small": 1000, "Medium": 1500, "Large": 2000}},
            2: {"name": "BBQ Pizza", "sizes": {"Small": 1000, "Medium": 1600, "Large": 2200}},
            3: {"name": "Fajita Pizza", "sizes": {"Small": 1000, "Medium": 1700, "Large": 2500}}
        }
        self.total_bill = 0
    def display_menu(self):
        for key, item in self.menu.items():
            print(f"{key}. {item['name']}")
            for size, price in item["sizes"].items():
                print(f"   {size} = {price}")
    def order_pizza(self):
        choice = int(input("Select Pizza 1-3: "))
        if choice in self.menu:
            print("Available Sizes: Small, Medium, Large")
            size = input("Enter size: ").capitalize()
            if size in self.menu[choice]["sizes"]:
                quantity = int(input("Enter quantity: "))
                price = self.menu[choice]["sizes"][size]
                bill = price * quantity
                self.total_bill += bill
                print(f"Added to cart: {quantity} {size} {self.menu[choice]['name']} = {bill}")
            else:
                print("Invalid Size no")
        else:
            print("Invalid Choice")
    def apply_discount(self):
        discount = 0
        if self.total_bill >= 5000 and self.total_bill <= 10000:
            discount = self.total_bill * 0.10
        elif self.total_bill >= 20000:
            discount = self.total_bill * 0.20
        final_bill = self.total_bill - discount
        print("Total Bill:", self.total_bill)
        print("Discount:", discount)
        print("Final Bill:", final_bill)
shop = FastFoodOrderMenu()
shop.display_menu()
while True:
    shop.order_pizza()
    more = input("Do you want to order more? yes or no: ").lower()
    if more != "yes":
        break
shop.apply_discount()
```
